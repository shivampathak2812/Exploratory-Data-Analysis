```python
#univariate

sns.countplot(x='category', data=df)
plt.title('Category')
plt.show()
```

```python
sns.histplot(x='opening_price', data=df)
plt.title('Opening Price')
plt.show()
```

```python
sns.stripplot(x='volume', data=df)
plt.title('strip plot')
plt.show()
```

```python
sns.violinplot(x='closing_price', data=df)
plt.title('violin')
plt.show()
```

```python
#bivariate
sns.scatterplot(x='opening_price', y='highest_price', hue='category', data=df)
plt.title('scatter plot')
plt.show()
```

```python
sns.regplot(x='closing_price', y='lowest_price', data=df)
plt.title('regression plot')
plt.show()
```

```python
sns.jointplot(x='opening_price', y='volume', data=df, kind='hex')
plt.title('joint plot')
plt.show()
```

```python
df.info()
```

```python
sns.violinplot(x='category', y='volume', data=df)
plt.title('violin plot')
plt.show()
```

```python
sns.swarmplot(y='highest_price', x='category', data=df)
plt.title('swarmplot')
plt.show()
```

```python
#category_color=['#0000FF', '#FF0000 ']
```

```python
sns.swarmplot(y='highest_price', x='category', data=df, palette='coolwarm')
plt.title('swarmplot')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df)
plt.title('box plot')
plt.show()
```

```python
sns.pairplot(df)
plt.title('pair plot')
plt.show()
```

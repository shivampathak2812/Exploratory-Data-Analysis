```python
#univariate

sns.countplot(x='Temperature', data=df)
plt.title('distribution of temoerateur')
plt.show()
```

```python
sns.barplot(x='Temperature', data=df)
plt.title('distribution of temperature')
plt.show()
```

```python
sns.histplot(x='Fuel_Price', data=df)
plt.title('distribution of fuel price')
plt.show()
```

```python
#bivariate

sns.barplot(x='Temperature', hue='Fuel_Price', data=df)
plt.title('dustibution of fuel price')
plt.show()
```

```python
sns.boxplot(x='Temperature', hue='Fuel_Price', data=df)
plt.title('boxplot of temperature')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df)
```

```python
#check the outlier for each column

for col in ['Store', 'Holiday_Flag', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment']:
  sns.boxplot(x=df[col])
  plt.title('boxplot')
  plt.show()
```

```python
sns.pairplot(data=df, kind='reg', diag_kind='kde')
plt.title('pairplot')
plt.show()
```

```python
sns.scatterplot(x='Temperature', y='Holiday_Flag', data=df)
plt.title('scatterplot')
plt.show()
```

```python
sns.scatterplot(x='Temperature', y='Fuel_Price', data=df, kind='reg')
plt.title('scatterplot')
plt.show()
```

```python
sns.regplot(x='Temperature', y='Fuel_Price', data=df)
plt.title('regression plot')
plt.show()
```

```python
correlation_matrix=df.corr()
print(correlation_matrix)
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('correlation matrix')
plt.show()
```

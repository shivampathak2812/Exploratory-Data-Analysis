```python
#numerical varible
# Analyze the column (a numerical variable)
sns.histplot(df['Over_Id'],kde=True)
plt.title('distribution of over id')
plt.show()
```

```python
plt.figure(figsize=(20,13))
sns.boxplot(data=df)
plt.title('boxplot of all variables')
plt.show()
```

```python
sns.boxplot(x='Batsman_Scored', data=df)
plt.title('Box plot of batsman scored')
plt.show()
```

```python
#bivariate analysis
#numerical and numerical variable

sns.scatterplot(x='Over_Id' , y='Ball_Id', data=df)
plt.title('scatter plot of overid and ballid')
plt.show()
```

```python
correlation_matrix=df.select_dtypes(include='number').corr()
print(correlation_matrix)
```

```python
#visualize the correlation using heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('correlation')
plt.show()
```

```python
#numerical and categorical
sns.boxplot(x='Batsman_Scored', hue='Over_Id', data=df)
plt.title("boxplot of batsman scored and over id")
plt.show()
```

```python
sns.boxplot(x='Over_Id', y='Ball_Id', hue='Batsman_Scored', data=df)
plt.title
```

```python
#Multivariate analysis
sample_df=df.sample(2000, random_state=42)
sns.pairplot(sample_df,
             hue='Batsman_Scored')
plt.title('Pair plot of dataset')
plt.show()
```

Exploratory Data Analysis (EDA) with Python: A Step-by-Step Guide

Exploratory Data Analysis (EDA) is a crucial initial step in the data science pipeline. It involves summarizing, visualizing, and understanding the main characteristics of a dataset. EDA helps data scientists and analysts to identify patterns, detect outliers, test hypotheses, and check assumptions before applying more complex statistical or machine learning techniques. Python, with its rich ecosystem of libraries such as Pandas, NumPy, Matplotlib, and Seaborn, provides a powerful environment for performing EDA efficiently. In this blog post, we will take you through a step-by-step guide on how to perform EDA using Python. Weâ€™ll cover the fundamental concepts, usage methods, common practices, and best practices.

```python
#Step 1 of importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#Step 2 loading of dataset

```python
#step2 Loading of dataset

iris=sns.load_dataset('iris')
```

# Step 4 data Cleaning

```python
#checking for missing value, outliers and duplicate
#checking the  missing value

iris.isnull().sum()
```

```python
#we dont have missing value
#iris=iris.dropna()
```

```python
#checking duplicated
print(iris.duplicated().sum())
```

```python
iris=iris.drop_duplicates()  #dropping duplicate values
```

```python
print(iris.duplicated().sum())  #check for dupliactes
#data is free from duplicated values
```

# Step 5 Univariate Analysis

```python
#based on data we choose different plots - numerical (hist and dist) and categorical (countplot)
#Univariate analysis focuses on analyzing a single variable at a time
```

```python
#Categorical data
species_count=iris['species'].value_counts()
```

```python
species_count
#species is a categorical data in iris dataset amd it consist of unique categories i.e., setosa, versicolor, virginica
```

```python
# Visualize the distribution of the'species' column

sns.countplot(data=iris, x='species')
plt.title('distribution of data')
plt.show()
```

```python
# Now We well find the numerical variable and we will check it through over hist plot and box plot

sns.histplot(iris['sepal_length'], kde=True)
plt.title('Distribution of sepal length')
plt.show()
```

```python
#now i will check numerical distribution by using boxplot
sns.boxplot(x='sepal_length', data=iris)
plt.title('boxplot chart')
plt.show()
```

```python
sns.boxplot(data=iris)
```

```python
sns.boxplot(x='species', data=iris)
```

# Step 6 Bivariate Anlysis

```python
#Bivariate analysis examines the relationship between two variables.

#2 Numerical variable
# Scatter plot between'sepal_length' and 'petal_length'

sns.scatterplot(x='sepal_length', y='petal_length', data=iris)
plt.title('Scatter plot')
plt.show()
```

```python
#now we calculate the correlation between 2 numerical value
# matrix=iris.corr()
# print(matrix)
```

```python
iris=iris.drop('species',axis=1)
```

```python
iris
```

```python
correlation_matrix=iris.corr()
print(correlation_matrix)
```

```python
# Visualize the correlation matrix as a heatmap

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation matrix')
plt.show()
```

# Step 7 Multivariate Analysis

```python
sns.pairplot(iris,
             kind='reg',   #add regression
             diag_kind='kde')  #kde plot on the diagonal
plt.title('Pairplot')
plt.show()
```

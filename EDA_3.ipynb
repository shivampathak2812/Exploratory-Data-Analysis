```python
#univariate

sns.countplot(x='hypertension', data=df)
plt.title('distibution')
plt.show
plt.legend()
```

```python
sns.countplot(x='Residence_type', data=df)
plt.title('Distribution of data')
plt.show()
```

```python
sns.countplot(x='work_type', data=df)
plt.title('distribution')
plt.show()
```

```python
sns.countplot(x='age', hue='work_type', data=df)
plt.title("distirbution of age, work and residence type")
plt.show()
```

```python
sns.histplot(x='ever_married', hue='work_type', data=df)
plt.title("distibution of ever married people")
plt.show()
```

```python
sns.countplot(x='smoking_status', hue='hypertension', data=df)
plt.title("distrivution of smoking people")
plt.show()
```

```python
sns.histplot(x='heart_disease', bins=20, kde=True, data=df)
plt.title("distibution of heart disease")
plt.show()
```

```python
sns.boxplot(x='work_type', data=df)
plt.title("distibution of work type")
plt.show()
```

```python
plt.figure(figsize=(10,9))
sns.boxplot(data=df)
```

```python
#bivariate

sns.scatterplot(x='age', y='work_type', data=df)
plt.title("distibution of age, work and residence type")
plt.show()
```

```python
sns.boxplot(x='gender', y='age', data=df)
plt.title("distibution of gender and age")
plt.show()
```

```python
sns.violinplot(x='smoking_status', y='age', data=df)
plt.title("distibution of smoking status and age")
plt.show()
```

```python
#multivariate

sns.scatterplot(x='age', y='work_type', hue='Residence_type', data=df)
plt.title("distibution of age, work and residence type")
plt.show()
```

```python
sns.pairplot(df, hue='work_type')
plt.title("distibution of gender and age")
plt.show()
```

```python
sns.boxplot(x='age', data=df)
plt.title("outlier detection")
plt.show()
```

```python
sns.heatmap(x='avg.glucose_level',data=df, annot=True, cmap='coolwarm')
plt.title("heatmap")
plt.show()
```
